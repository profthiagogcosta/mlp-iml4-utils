# Setup

```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn import metrics

sample_url = "Mall_Customers.csv"
sample_data = pd.read_csv(sample_url)

data = sample_data.iloc[:, 3:5].values
```

# Ward and euclidian and K fixed

```python
plt.figure(figsize=(10, 7))
plt.title("Ward linkage with euclidean metrics Dendograms")
dend = sch.dendrogram(sch.linkage(data, method='ward', metric='euclidean'))

euclidian_cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
euclidian_cluster.fit_predict(data)

plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=euclidian_cluster.labels_, cmap='rainbow')
```

```python
silhoutte_metric = metrics.silhouette_score(data, euclidian_cluster.labels_)
print("silhoutte_metric: ", silhoutte_metric)
calinski_harabasz_metric = metrics.calinski_harabasz_score(data, euclidian_cluster.labels_)
print("calinski_harabasz_metric: ", calinski_harabasz_metric)
davies_bouldin_metric = metrics.davies_bouldin_score(data, euclidian_cluster.labels_)
print("davies_bouldin_metric: ", davies_bouldin_metric)
```

```python
plt.figure(figsize=(10, 7))
plt.title("Ward linkage with euclidean metrics Dendograms")
dend = sch.dendrogram(sch.linkage(data, method='ward', metric='euclidean'))

euclidian_cluster = AgglomerativeClustering(n_clusters=None, affinity='euclidean', linkage='ward', distance_threshold=100)
euclidian_cluster.fit_predict(data)

plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=euclidian_cluster.labels_, cmap='rainbow')

euclidian_cluster.n_clusters_
```
